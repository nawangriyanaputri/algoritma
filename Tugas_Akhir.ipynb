```python
class Kursus:
    def __init__(self, nama_kursus, kode_kursus, durasi, kapasitas):
        self.nama_kursus = nama_kursus
        self.kode_kursus = kode_kursus
        self.durasi = durasi  # dalam minggu
        self.kapasitas = kapasitas
        self.peserta_terdaftar = 0

    def tambah_peserta(self):
        if self.peserta_terdaftar < self.kapasitas:
            self.peserta_terdaftar += 1
            print(f"Peserta berhasil ditambahkan ke kursus {self.nama_kursus}.")
        else:
            print(f"Kursus {self.nama_kursus} sudah penuh.")

    def info_kursus(self):
        return (f"Nama Kursus: {self.nama_kursus}"
                f"Kode Kursus: {self.kode_kursus}"
                f"Durasi: {self.durasi} minggu"
                f"Kapasitas: {self.kapasitas}"
                f"Peserta Terdaftar: {self.peserta_terdaftar}")



daftar_kursus = []
kode_ke_kursus = {}


def tambah_kursus(nama_kursus, kode_kursus, durasi, kapasitas):
    try:
        if not isinstance(durasi, int) or not isinstance(kapasitas, int):
            raise TypeError("Durasi dan kapasitas harus berupa angka.")
        kursus = Kursus(nama_kursus, kode_kursus, durasi, kapasitas)
        daftar_kursus.append(kursus)
        kode_ke_kursus[kode_kursus] = kursus
        print(f"Kursus '{nama_kursus}' berhasil ditambahkan.")
    except TypeError as e:
        print(f"Kesalahan: {e}")


def tampilkan_semua_kursus():
    if not daftar_kursus:
        return "Tidak ada kursus yang tersedia."
    return "\n".join([kursus.info_kursus() for kursus in daftar_kursus])


def cari_kursus(kode):
    try:
        kursus = kode_ke_kursus[kode]
        print(kursus.info_kursus())
    except KeyError:
        print(f"Kursus dengan kode '{kode}' tidak ditemukan.")


def tambah_peserta_ke_kursus(kode):
    try:
        kursus = kode_ke_kursus[kode]
        kursus.tambah_peserta()
    except KeyError:
        print(f"Kursus dengan kode '{kode}' tidak ditemukan.")
    except ValueError as e:
        print(f"Kesalahan: {e}")



if __name__ == "__main__":

    tambah_kursus("Bahasa Indonesia", "BI035", 5, 14)
    tambah_kursus("Bahasa Inggris", "BAS27", 6, 12)


    print("\nDaftar Semua Kursus:")
    tampilkan_semua_kursus()


    print("\nMencari Kursus dengan Kode 'BI035':")
    cari_kursus("BI035")
    print("\nMencari Kursus dengan Kode 'INVALID':")
    cari_kursus("INVALID")


    print("\nMenambah Peserta ke Kursus 'BI035':")
    tambah_peserta_ke_kursus("BI035")
    print("\nMenambah Peserta ke Kursus dengan Kode 'INVALID':")
    tambah_peserta_ke_kursus("INVALID")


    print("\nMenambah Peserta hingga Kapasitas Penuh:")
    for _ in range(14):
        tambah_peserta_ke_kursus("BI035")
```
