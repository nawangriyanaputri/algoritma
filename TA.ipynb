```python
class Kursus:
    def __init__(self, nama_kursus, kode_kursus, durasi, kapasitas):
        self.nama_kursus = nama_kursus
        self.kode_kursus = kode_kursus
        self.durasi = durasi  # dalam minggu
        self.kapasitas = kapasitas
        self.peserta_terdaftar = 0

    def tambah_peserta(self, jumlah):
        if self.peserta_terdaftar + jumlah <= self.kapasitas:
            self.peserta_terdaftar += jumlah
            print(f"{jumlah} peserta berhasil ditambahkan ke kursus {self.nama_kursus}.")
        else:
            print(f"Tidak dapat menambah {jumlah} peserta, kapasitas kursus {self.nama_kursus} hanya {self.kapasitas - self.peserta_terdaftar} peserta lagi.")

    def info_kursus(self):
        return (f"Nama Kursus: {self.nama_kursus}\n"
                f"Kode Kursus: {self.kode_kursus}\n"
                f"Durasi: {self.durasi} minggu\n"
                f"Kapasitas: {self.kapasitas}\n"
                f"Peserta Terdaftar: {self.peserta_terdaftar}")

# Daftar kursus dan kode untuk mencari kursus berdasarkan kode
daftar_kursus = []
kode_ke_kursus = {}

# Menambahkan kursus
def tambah_kursus(nama_kursus, kode_kursus, durasi, kapasitas):
    try:
        if not isinstance(durasi, int) or not isinstance(kapasitas, int):
            raise TypeError("Durasi dan kapasitas harus berupa angka.")
        kursus = Kursus(nama_kursus, kode_kursus, durasi, kapasitas)
        daftar_kursus.append(kursus)
        kode_ke_kursus[kode_kursus] = kursus
        print(f"Kursus '{nama_kursus}' berhasil ditambahkan.")
    except TypeError as e:
        print(f"Kesalahan: {e}")

def tambah_kursus_otomatis():
    kursus_list = [
        {"Nama": "Bahasa Indonesia", "Kode": "BI035", "Durasi": 5, "Kapasitas": 14},
        {"Nama": "Bahasa Inggris", "Kode": "BAS27", "Durasi": 6, "Kapasitas": 12},
        {"Nama": "Dasar - Dasar Python", "Kode": "DDP08", "Durasi": 8, "Kapasitas": 15}
    ]

    for kursus_data in kursus_list:
        kursus = Kursus(kursus_data["Nama"], kursus_data["Kode"], kursus_data["Durasi"], kursus_data["Kapasitas"])
        daftar_kursus.append(kursus)
        kode_ke_kursus[kursus_data["Kode"]] = kursus
    print("Kursus-kursus otomatis telah ditambahkan.")

# Menampilkan semua kursus
def tampilkan_semua_kursus():
    if not daftar_kursus:
        print("Tidak ada kursus yang tersedia.")
    else:
        for kursus in daftar_kursus:
            print(kursus.info_kursus())

# Mencari kursus berdasarkan kode
def cari_kursus(kode):
    try:
        kursus = kode_ke_kursus[kode]
        print(kursus.info_kursus())
    except KeyError:
        print(f"Kursus dengan kode '{kode}' tidak ditemukan.")

# Menambah peserta ke kursus
def tambah_peserta_ke_kursus(kode):
    try:
        kursus = kode_ke_kursus[kode]
        jumlah = int(input("Jumlah peserta yang ingin ditambahkan: "))
        kursus.tambah_peserta(jumlah)
    except KeyError:
        print(f"Kursus dengan kode '{kode}' tidak ditemukan.")
    except ValueError:
        print("Jumlah peserta harus berupa angka.")

# Input pengguna
def menu():
    while True:
        print("\n==== Menu Kursus ====")
        print("1. Tambah Kursus")
        print("2. Tampilkan Semua Kursus")
        print("3. Cari Kursus")
        print("4. Tambah Peserta ke Kursus")
        print("5. Keluar")

        pilihan = input("Pilih menu (1-5): ")

        if pilihan == "1":
            nama_kursus = input("Masukkan nama kursus: ")
            kode_kursus = input("Masukkan kode kursus: ")
            try:
                durasi = int(input("Masukkan durasi kursus (dalam minggu): "))
                kapasitas = int(input("Masukkan kapasitas kursus: "))
                tambah_kursus(nama_kursus, kode_kursus, durasi, kapasitas)
            except ValueError:
                print("Durasi dan kapasitas harus berupa angka.")

        elif pilihan == "2":
            print("\nDaftar Semua Kursus:")
            tampilkan_semua_kursus()

        elif pilihan == "3":
            kode = input("Masukkan kode kursus yang ingin dicari: ")
            cari_kursus(kode)

        elif pilihan == "4":
            kode = input("Masukkan kode kursus untuk menambah peserta: ")
            tambah_peserta_ke_kursus(kode)

        elif pilihan == "5":
            print("Program selesai.")
            break
        else:
            print("Pilihan tidak valid. Silakan pilih antara 1 hingga 5.")

if __name__ == "__main__":
    tambah_kursus_otomatis()
    menu()
```
